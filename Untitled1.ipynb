```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix, accuracy_score, classification_report
```

```python
# Set up working directory and suppress warnings
warnings.filterwarnings("ignore")
```

```python
# Load and inspect the dataset
df = pd.read_csv("data.csv")
print("Initial shape of the dataset:", df.shape)
```

```python
df.head()
```

```python
# Data Cleaning
df_cleaned = df.drop_duplicates(subset=["Customer_ID"])
print("Shape of the dataset after removing duplicates:", df_cleaned.shape)
df_cleaned.drop(["ID", "Customer_ID", "Name", "SSN", "Type_of_Loan"], axis=1, inplace=True)
print("Shape of the dataset after dropping columns:", df_cleaned.shape)
```

```python
# Data Preprocessing
df_encoded = pd.get_dummies(df_cleaned)
df_encoded.drop(["Occupation_Writer", "Credit_Mix_Standard", "Payment_of_Min_Amount_NM",
                 "Payment_Behaviour_Low_spent_Small_value_payments", "Credit_Score_Standard"], axis=1, inplace=True)
```

```python
df_encoded.head()
```

```python
# Splitting Data
dep = df_encoded["Monthly_Balance"]
ind = df_encoded.drop(["Monthly_Balance"], axis=1)
x_train, x_test, y_train, y_test = train_test_split(ind, dep, train_size=0.75, random_state=42)
```

```python
# Linear Regression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
linear_pred = linear_model.predict(x_test)
r2 = r2_score(y_test, linear_pred)
mse = mean_squared_error(y_test, linear_pred)
```

```python
residuals = y_test - linear_pred
plt.hist(residuals, bins=30, color='blue', alpha=0.7)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```

```python
plt.scatter(y_test, linear_pred, color='blue', alpha=0.7)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Scatter Plot: Actual vs. Predicted')
plt.show()
```

```python
plt.hist(y_test, bins=30, color='blue', alpha=0.7, label='Actual')
plt.hist(linear_pred, bins=30, color='orange', alpha=0.5, label='Predicted')
plt.xlabel('Monthly Balance')
plt.ylabel('Frequency')
plt.title('Distribution of Actual and Predicted Monthly Balance')
plt.legend()
plt.show()
```

```python
print("Linear Regression Model:")
print("R-Square Value:", round(r2, 4))
print("Mean Squared Error:", mse)
```
